# Regresión lineal simple python vs orange


```python
import pandas as pd
import sklearn.linear_model as sklm
import seaborn as sns
```

```python
dfOrange = pd.read_csv("regresion lineal simple.csv")
dfPython = pd.read_csv("automovilesEDA-curso.csv")
```

# Regresión lineal simple python 

```python
# se define la varible predictora (x)y la varible objetivo (y)
x = dfPython[["mpgCarretera"]]
y = dfPython["PRECIO"]
# crear varible para la regresión lineal
lm = sklm.LinearRegression()
# ingresamos los parametros
lm.fit(x, y)
# se obtiene la predicción
prediccion = lm.predict(x)
prediccion[0:10]
# comparar datos
arrayMultiple = {"Y": y, "Predicción": prediccion.round(
    2), "X": x.to_numpy().flatten()}
dfPython = pd.DataFrame(arrayMultiple)
dfPython.sample(15)
```

# Regresión lineal simple Orange

```python
# Visualizar los datos independientes de la prediccion
dfOrange
```

# Limpiar los datos

```python
# Eliminar datos inservibles
elim = dfOrange["Linear Regression"].drop([0, 1], axis=0)
elim
```

# Comparar datos <b> Orange </b> con datos <b> Python </b>

```python
# Visualizar la prediccion del precio de python
pi = sns.distplot(prediccion, hist=False, color="r", label="Valor Real")
pi
```

```python
# Visualizar la prediccion de orange
sns.distplot(elim, hist=False, color="y", label="Valor Real")
```

# Comparar la predicciones de orange y python

```python
# Visualizar la comparacion entre orange y python
pi = sns.distplot(prediccion, hist=False, color="r", label="Valor Real")
sns.distplot(elim, hist=False, color="b", label="Valor Real", ax=pi)
```
